```python
import numpy as np, pandas as pd
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.figsize':(9,7), 'figure.dpi':120})
```

```python
df = pd.read_csv('May02_market_data.csv')
```

```python
df = df[df['QUOTIM']>'04:00:00']
```

```python
df
```

```python
fig, axes = plt.subplots(3, 2, sharex=True)
axes[0, 0].plot(df.BID); axes[0, 0].set_title('Original Series')
plot_acf(df.BID, ax=axes[0, 1])

axes[1, 0].plot(df.BID.diff()); axes[1, 0].set_title('1st Order Differencing')
plot_acf(df.BID.diff().dropna(), ax=axes[1, 1])

axes[2, 0].plot(df.BID.diff().diff()); axes[2, 0].set_title('2nd Order Differencing')
plot_acf(df.BID.diff().diff().dropna(), ax=axes[2, 1])

```
